```python
import pandas as pd
```

```python
ds = pd.read_csv('/content/drive/MyDrive/DACN/CTi/NVD_all_csv.csv', delimiter=';')
ds.head()
```

```python
unique_POS = ds['POS'].unique()
unique_POS = sorted(unique_POS)
print(unique_POS)
```

```python
unique_tag = ds['tag'].unique()
unique_tag = sorted(unique_tag)
print(unique_tag)
```

```python
pos_2_id = {}
for i, pos in enumerate(unique_POS):
  pos_2_id[pos] = i
id_2_pos = {i: pos for pos, i in pos_2_id.items()}
print(pos_2_id)
print(id_2_pos)
```

```python
tag_2_id = {}
for i, tag in enumerate(unique_tag):
  tag_2_id[tag] = i
id_2_tag = {i: tag for tag, i in tag_2_id.items()}
print(tag_2_id)
print(id_2_tag)
```

```python
ds['POS'] = ds['POS'].map(lambda x: pos_2_id[x])
ds['tag'] = ds['tag'].map(lambda x: tag_2_id[x])
ds.head()
```

```python
result_df = ds.groupby('sentence_idx').agg({'words': list, 'POS': list, 'tag': list}).reset_index()
result_df.head()
```

```python
len(result_df.iloc[0]['words']) == len(result_df.iloc[0]['tag'])
```
